```python
import cv2
vidcap = cv2.VideoCapture('resources/eyes/Vector graphics Eyes animation.mp4')
success,image = vidcap.read()
count = 0
success = True
stack = []
while success:
  success,image = vidcap.read()
  #image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
  image = image[230:490, 700:1150]
  stack.append(image)
  
  
```

```python
stack = np.array(stack)
```

```python
import numpy as np
import cv2
size = 260,450
fps = 25
out = cv2.VideoWriter('resources/eyes/move.mp4', cv2.VideoWriter_fourcc(*'mp4v'), fps, (size[1], size[0]), False)
np.save( 'resources/eyes/move.npy',stack[30:250])
for x in stack[30:250]:
    out.write(x)
out.release()
```

```python
import numpy as np
import cv2
size = 260,450
fps = 25
out = cv2.VideoWriter('resources/eyes/blink.mp4', cv2.VideoWriter_fourcc(*'mp4v'), fps, (size[1], size[0]), False)

for x in stack[0:30]:
    out.write(x)
for x in np.flip(stack[0:30],axis=0):
    out.write(x)
np.save( 'resources/eyes/blink.npy',np.append(stack[0:30], np.flip(stack[0:30],axis=0),axis=0))


out.release()
```

```python
import numpy as np
import cv2
size = 260,450
fps = 25
out = cv2.VideoWriter('resources/eyes/stare.mp4', cv2.VideoWriter_fourcc(*'mp4v'), fps, (size[1], size[0]), False)

for x in stack[30:50]:
    out.write(x)

out.release()
np.save( 'resources/eyes/stare.npy',stack[30:50])
```
